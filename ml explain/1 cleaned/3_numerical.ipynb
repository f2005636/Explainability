# 3. numerical

```python
def num_var(x,y,z,i,j):
    dtree = tree.DecisionTreeClassifier(random_state=0, max_depth=i, min_samples_leaf=j)
    dtree = dtree.fit(df[[x]],df[y])
    
    df[z] = dtree.apply(df[[x]])
    gb = df.groupby(z).agg({'index':'count', x:['min','max'], y:'mean'}).reset_index()
    display(gb)
```

```python
num_var('Amount', 'Y', 'Amount_bin', 3, 42)
```

```python
num_var('Interest', 'Y', 'Interest_bin', 3, 42)
```

```python
num_var('IncomeTotal', 'Y', 'IncomeTotal_bin', 3, 42)
```

```python
num_var('LiabilitiesTotal', 'Y', 'LiabilitiesTotal_bin', 3, 42)
```

```python
num_var('DebtToIncome', 'Y', 'DebtToIncome_bin', 3, 42)
```

# 4. groups

```python
def cat_var(x,y):
    gb = df.groupby(x).agg({'index':'count',y:'mean'}).reset_index()
    display(gb)
```

```python
df['Amount_bin'] = np.where(df['Amount']<=530.35,0,
                            np.where(df['Amount']<=532,-1,
                                     np.where(df['Amount']<=2100,0,
                                              np.where(df['Amount']<=2125,1,0))))
df['Interest_bin'] = np.where(df['Interest']<=48.22,-1,
                              np.where(df['Interest']<=79.87,0,1))
df['LoanDuration_bin'] = np.where(df['LoanDuration']<=30,-1,1)
df['MonthlyPayment_bin'] = np.where(df['MonthlyPayment']<=0,1,
                                    np.where(df['MonthlyPayment']<=16.17,-1,0))
df['IncomeTotal_bin'] = np.where(df['IncomeTotal']<=304,-1,1)
df['LiabilitiesTotal_bin'] = np.where(df['LiabilitiesTotal']<=349.99,-1,
                                    np.where(df['LiabilitiesTotal']<=350.11,1,-1))
df['DebtToIncome_bin'] = np.where(df['DebtToIncome']<0.61,-1,
                                  np.where(df['DebtToIncome']<=62.35,0,1))
```

```python
cat_var('Amount_bin','Y')
cat_var('Interest_bin','Y')
cat_var('LoanDuration_bin','Y')
cat_var('MonthlyPayment_bin','Y')
cat_var('IncomeTotal_bin','Y')
cat_var('LiabilitiesTotal_bin','Y')
cat_var('DebtToIncome_bin','Y')
```
